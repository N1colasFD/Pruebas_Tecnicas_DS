```python
np.random.seed(123)

def clasificar_cliente(capital, edad, interes_simple):
    if edad > 60 and capital < 30000:
        return 'A'
    elif 25000 <= capital <= 30000 and 18 <= edad <= 70:
        return 'AA'
    elif capital > 35000 and interes_simple >= 0.06:
        return 'AAA'
    elif interes_simple == 0.04:
        return 'PPP'
    else:
        return 'SC'

n = 3000
capital = np.random.randint(20000, 40000, n)
interes_simple = np.random.uniform(0.03, 0.08, n)
edad = np.random.randint(18, 70, n)
saldo_3_anos_simple = capital * interes_simple * 3
saldo_3_anos_compuesto = capital * np.power(1 + interes_simple, 3)
rentabilidad = saldo_3_anos_compuesto - capital
tipo_cliente = [clasificar_cliente(c, e, i) for c, e, i in zip(capital, edad, interes_simple)]


df = pd.DataFrame({'capital': capital, 'interes_simple': interes_simple, 
                   'edad': edad, 'saldo_3_anos_simple': saldo_3_anos_simple, 
                   'saldo_3_anos_compuesto': saldo_3_anos_compuesto,
                   'rentabilidad':rentabilidad,
                   'tipo_cliente': tipo_cliente})

df[['capital', 'saldo_3_anos_simple', 'saldo_3_anos_compuesto', 'rentabilidad']] = \
    df[['capital', 'saldo_3_anos_simple', 'saldo_3_anos_compuesto', 'rentabilidad']].round(2)

fecha_inicial_min = pd.Timestamp(datetime(2022, 1, 1))
fecha_inicial_max = pd.Timestamp(datetime(2022, 12, 31))
diferencia = fecha_inicial_max - fecha_inicial_min
fechas_inicio = [fecha_inicial_min + pd.Timedelta(days=random.randint(0, diferencia.days)) for i in range(3000)]
df['fecha_inicio'] = fechas_inicio
df['fecha_fin'] = df['fecha_inicio'] + timedelta(days=3*365)

df.sample(n = 50)
```

```python
def agregar_registro(df, index):
    capital = int(input("Ingrese el capital: "))
    interes_simple = float(input("Ingrese el inter√©s simple: "))
    edad = int(input("Ingrese la edad: "))
    saldo_3_anos_simple = capital * (1 + 3 * interes_simple)
    saldo_3_anos_compuesto = capital * (1 + interes_simple) ** 3
    rentabilidad = saldo_3_anos_compuesto - capital
    fecha_inicio = pd.Timestamp(datetime.now())
    fecha_fin = fecha_inicio + timedelta(days=3*365)
    tipo_cliente = clasificar_cliente(capital, edad, interes_simple)
    df.loc[index] = [capital, interes_simple, edad, saldo_3_anos_simple, saldo_3_anos_compuesto, fecha_inicio, fecha_fin, tipo_cliente]

df = pd.DataFrame(columns=['capital', 'interes_simple', 'edad', 'saldo_3_anos_simple', 'saldo_3_anos_compuesto','fecha_inicio', 'fecha_fin', 'tipo_cliente'])
def agregar_multiples_registros(df, num_registros):
    df['fecha_inicio'] = None
    for i in range(0, 1 + num_registros):
        agregar_registro(df, i)
agregar_multiples_registros(df, 2)
```
